# **Kapitel 1: Business Understanding**  

## **Einleitung**  
Streaming-Dienste wie Netflix und Disney+ bieten eine riesige Auswahl an Filmen und Serien. Doch genau diese Vielfalt stellt viele Nutzer vor ein Problem: **Welche Show soll ich mir ansehen?** Die Entscheidungsfindung kann durch zu viele Optionen überfordernd sein, was oft zu langem Suchen oder sogar zur Frustration führt. Das Ziel dieses Projekts ist es, eine **Show-Empfehlung** basierend auf den individuellen Vorlieben der Nutzer bereitzustellen, um die Entscheidungszeit zu minimieren und das Seherlebnis zu verbessern.  

## **1.1 Betriebswirtschaftlichen Problemstellung**  
Das Hauptproblem ist die **Überforderung durch die große Auswahl**. Nutzer möchten sich einfach zurücklehnen und eine Show genießen, ohne lange suchen zu müssen. Die Empfehlung soll daher auf Basis von Nutzerpräferenzen erfolgen, um eine schnellere und gezieltere Auswahl zu ermöglichen.  

**Operationalen und betriebswirtschaftlichen Zielkriterien:**  
- Bereitstellung einer **Top-10-Liste** basierend auf Nutzerpräferenzen.  
- Reduzierung der Suchzeit durch gezielte Vorschläge.  
- Verbesserung der Nutzererfahrung durch Personalisierung.  
- Sicherstellung der Verfügbarkeit der empfohlenen Inhalte im abonnierten Streaming-Dienst.  

## **1.2 Situationsbewertung**  
### **Datenquellen**  
- Filme und Serien von **Netflix und Disney+** mit Metadaten wie Genre, Altersfreigabe und Dauer.  
- Nutzereingaben (Alter, Genre-Präferenzen, Streaming-Abonnement, Anlass).  

### **Software & Tools**  
- **Programmiersprachen:** Python für Datenanalyse und Modellierung.  
- **Bibliotheken:** Pandas, Scikit-Learn, Matplotlib für Analysen und Visualisierungen.  
- **Datenbank:** Speicherung von Nutzereingaben für Personalisierung.  
- **Frontend:** Web-UI zur Eingabe von Präferenzen.  

### **Mögliche Risiken**  
- **Datenqualität:** Fehlende oder unvollständige Metadaten zu Filmen und Serien.  
- **Nutzerengagement:** Nutzer geben ungenaue oder zu wenige Informationen an.   
- **Algorithmische Herausforderungen:** Unterschiedliche Geschmäcker erschweren die Vorhersage.  

## **1.3 Bestimmung analytischer Ziele**  
Das Ziel ist es, eine **personalisierte Empfehlung für Filme und Serien** zu entwickeln. Dazu werden folgende Fragen analysiert:  
- Welche Genres bevorzugt der Nutzer?  
- Hat der Nutzer eine Präferenz für Filme oder Serien?  
- Welche Altersfreigaben sind relevant?  
- Bevorzugt der Nutzer Inhalte einer bestimmten Plattform?  
- Hat der Nutzer Lieblingsschauspieler oder -regisseure?  
- Welcher Anlass ist für die Auswahl relevant (z. B. Mädelsabend, Halloween, Date Night)?  

**Erfolgskriterium:**  
- Eine **schnellere und zufriedenstellende Empfehlung**, die basierend auf Nutzerpräferenzen eine passende Show vorschlägt und damit die Entscheidungszeit minimiert.  

## **1.4 Erstellung des Projektplans**  
### **Projektphasen & Zeitrahmen**  
| Phase | Aufgaben | Zeitspanne |
|--------|--------------------------|--------------|
| Business Understanding | Definition der Problemstellung & Ziele | 10% |
| Data Understanding | Datenanalyse & Qualitätssicherung | 20% |
| Data Preparation | Datenbereinigung & Feature Engineering | 30% |
| Modeling | Entwicklung des Empfehlungsmodells | 20% |
| Evaluation | Modellbewertung & Optimierung | 10% |
| Deployment | Integration in die Web-Oberfläche | 10% | 

### **Risiken & Maßnahmen**  
| Risiko | Mögliche Auswirkungen | Gegenmaßnahmen |
|--------|------------------------|----------------|
| Fehlende Daten | Unvollständige Empfehlungen | Ergänzung durch externe Datenquellen |
| Komplexe Nutzerwünsche | Erschwerte Modellierung | Gewichtung von Prioritäten |
| Skalierbarkeit | Serverüberlastung | Cloud-Hosting & Caching |
| Algorithmische Ungenauigkeiten | Falsche Empfehlungen | Modell-Feinabstimmung & Feedbacksystem |

Durch diesen strukturierten Ansatz wird sichergestellt, dass die Empfehlungen präzise, effizient und nutzerfreundlich sind.
