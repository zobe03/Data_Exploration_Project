Diese Phase dient dem Analysten, bestehende Zusammenhänge aus den Daten zu erkennen, eventuelle Qualitätsmängel der Daten festzustellen oder interessante Teilmengen zu identifizieren, um eine Hypothese über die Daten aufzustellen. Die Phase besteht aus folgenden vier Schritten:)

## **2.1 Daten sammeln**
In diesem Schritt wurden die benötigten Daten für die Analyse beschafft. Zwei Hauptdatenquellen wurden ausgewählt:
- **Netflix-Titelliste**: Enthält Informationen zu 8.807 Filmen und Serien auf Netflix, inklusive Genre, Altersfreigabe, Erscheinungsjahr, Regisseur und Hauptdarsteller.
- **Disney+ Titelliste**: Enthält 1.450 ähnliche Metadaten zu Disney+-Inhalten.

Während der Datenbeschaffung traten folgende Herausforderungen auf:
- Unvollständige oder fehlende Angaben zu Regisseuren und Schauspielern (Netflix: 2.634 fehlende Regieangaben, Disney+: 473 fehlende Regieangaben).
- Unterschiedliche Kategorisierung von Genres zwischen den Plattformen (Netflix: 514 Genres, Disney+: 329 Genres).
- Variierende Altersfreigaben (Netflix: 17 unterschiedliche Ratings, Disney+: 9 unterschiedliche Ratings).
- Fehlplatzierte Werte in der Netflix-Datenbank, wo einige Einträge in der Spalte 'rating' fälschlicherweise zur Spalte 'duration' gehören.

## **2.2 Daten beschreiben**
### **Allgemeine Datenstruktur**
Beide Datensätze enthalten die folgenden Hauptfelder:
- **Titel**: Name des Films oder der Serie.
- **Typ**: Movie oder TV Show.
- **Regisseur**: Name des Regisseurs (falls vorhanden).
- **Schauspieler**: Liste der Hauptdarsteller.
- **Produktionsland**: Herkunftsland der Produktion (Netflix: 748 Länderangaben, Disney+: 89 Länderangaben).
- **Altersfreigabe**: Klassifizierung basierend auf Jugendschutzrichtlinien.
- **Dauer**: Länge des Films oder Anzahl der Staffeln bei Serien.
- **Genres**: Zugehörige Kategorien (z. B. Action, Drama, Comedy).
- **Beschreibung**: Kurze Inhaltsangabe.

### **Datenquantität und Formatierung**
| Plattform | Anzahl Einträge | Fehlende Werte (Regisseur, Cast, Land) | Format |  
|-----------|---------------|--------------------------------|--------|  
| Netflix   | 8.807         | 2.634 / 825 / 831             | CSV  |  
| Disney+   | 1.450         | 473 / 190 / 219               | CSV  |  

Die Daten sind in CSV-Format gespeichert und wurden in eine einheitliche Struktur überführt.

## **2.3 Untersuchung der Daten**
Erste Analysen zeigen:
- **Netflix hat mehr Serien als Disney+** (Netflix: 2.676 Serien, Disney+: 398 Serien).
- **Häufigste Genres auf beiden Plattformen sind Drama, Komödie und Action.**
- **Netflix hat eine breitere Spanne an Altersfreigaben, während Disney+ familienfreundlicher ist.**
- **Regisseure sind bei Netflix stärker vertreten als bei Disney+ (Netflix: 4.528, Disney+: 609).**
- **Falsche Werte in der Spalte 'rating' bei Netflix** wurden entdeckt (z. B. '66 min', '74 min', '84 min'). Diese Werte gehören in die Spalte 'duration'.

## **2.4 Bewertung der Daten**
Die Qualität der Daten wurde anhand folgender Kriterien geprüft:
- **Vollständigkeit**: Einige Felder enthalten fehlende Werte, insbesondere Regisseure und Schauspieler.
- **Konsistenz**: Unterschiedliche Bezeichnungen für Genres und Altersfreigaben zwischen Plattformen erfordern eine Standardisierung.
- **Relevanz**: Die vorhandenen Daten sind grundsätzlich gut geeignet, um Empfehlungen basierend auf Nutzerpräferenzen abzuleiten.

### **Erste Erkenntnisse und weitere Schritte**
- Bereinigung der Daten, insbesondere Korrektur der 'rating'- und 'duration'-Spalten in Netflix-Daten.
- Standardisierung der Genres und Altersfreigaben für eine bessere Vergleichbarkeit.
- Untersuchung der Genre-Cluster, um bessere Empfehlungen zu generieren.

Durch diese Analysen entsteht ein grundlegendes Verständnis für die Daten und deren Potenzial zur Entwicklung eines Empfehlungsmodells.
